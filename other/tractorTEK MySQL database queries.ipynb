```python
data = pd.read_sql_query('SELECT emp_id, quantity FROM prod_sales', engine)
data.sort_values(by='emp_id').head(35)
```

```python
byEmp = data.groupby('emp_id')
```

```python
## mean(avg) number of products sold grouped by employee id ##
byEmp.mean()
```

```python
byEmp.sum()
```

```python
warrdata = pd.read_sql_query('SELECT emp_id, quantity FROM warranty_sales',engine)
print(warrdata.head(10))
```

```python
warbyEmp = warrdata.groupby('emp_id')
```

```python
## mean(avg) number of warranties sold grouped by employee ID ##
warbyEmp.mean()
```

```python
## sum of warranties sold per employee ##
warbyEmp.sum()
```

```python
df = pd.read_sql_query('SELECT emp_id, esp_id, quantity FROM warranty_sales',engine)
print(df.head(25))
```

```python
## grouping by EMP ID, summing warranty sales quantities and only returning EMP ID that = emp234 ##
df.groupby('emp_id').sum().loc['EMP234']
```

```python
df.groupby('emp_id').sum().loc['EMP244']
```

```python
## max number of warranty sales per Extended Service Plan ##
df = pd.read_sql_query('SELECT esp_id, quantity FROM warranty_sales', engine)
df.groupby('esp_id').max()
```

```python
df = pd.read_sql_query('SELECT prod_id, quantity FROM prod_sales', engine)
df.groupby('prod_id').sum()
```

```python
## AVG number of products sold per Product ID ##
df.groupby('prod_id').mean()
```

```python
# max number of products sold per Product ID ##
df.groupby('prod_id').max()
```

```python
## least amount of products sold per Product ID ##

df.groupby('prod_id').min()
```

```python
df.groupby('prod_id').describe()
```

#### Concatenating tables ###

```python
df1 = pd.read_sql('prod_sales', engine)
```

```python
df2 = pd.read_sql('prod_prices', engine)
```

```python
df3 = pd.read_sql('warranty_sales', engine)
```

```python
df4 = pd.read_sql('warranty_prices', engine)
```

```python
pd.concat([df2,df4])
```

```python
## merging product sales table and warranty sales table on emp_id column ##

pd.merge(df1, df3, on=['emp_id'])
```
